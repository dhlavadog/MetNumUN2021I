# **ACTIVISION BLIZZARD**


```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2021, 1, 1)
        self.SetEndDate(2021, 7, 1)

        # Select stocks and add in DataNormalizationMode = Raw

        self.nflx = self.AddEquity("ATVI", Resolution.Minute)
        self.nflx.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        

    def OnData(self, data):
    # Capital base: 1 Millón de dolares (valores en cifras de miles)
        if not self.Portfolio.Invested:
            self.MarketOrder("ATVI", 100)
            
            #Depuración de precio promedio
            self.Debug(str(self.Portfolio["ATVI"].AveragePrice))
          
```

# **PORTAFOLIO**

Contruimos el portafolio con base a las 5 empresas anteriormente mencionadas.

### **Para Periodos de 6 meses:**

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        self.SetStartDate(2021, 1, 1)
        self.SetEndDate(2021, 7, 1)

        # Select stocks and add in DataNormalizationMode = Raw

        self.nflx = self.AddEquity("ATVI", Resolution.Minute)
        self.nflx.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.dis = self.AddEquity("MSFT", Resolution.Minute)
        self.dis.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.fb = self.AddEquity("GOOG", Resolution.Minute)
        self.fb.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.tsla = self.AddEquity("IBM", Resolution.Minute)
        self.tsla.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.ntdoy = self.AddEquity("ORCL", Resolution.Minute)
        self.ntdoy.SetDataNormalizationMode(DataNormalizationMode.Raw)
        

    def OnData(self, data):
    # Capital base: 1 Millón de dolares (valores en cifras de miles)
        if not self.Portfolio.Invested:
            self.MarketOrder("ATVI", 700)
            self.MarketOrder("MSFT", 100)
            self.MarketOrder("GOOG", 100)
            self.MarketOrder("IBM", 50)
            self.MarketOrder("ORCL", 50)
            
            #Depuración de precio promedio
            self.Debug(str(self.Portfolio["ATVI"].AveragePrice))
            self.Debug(str(self.Portfolio["MSFT"].AveragePrice))
            self.Debug(str(self.Portfolio["GOOG"].AveragePrice))
            self.Debug(str(self.Portfolio["IBM"].AveragePrice))
            self.Debug(str(self.Portfolio["ORCL"].AveragePrice))
```

Reiniciamos la iteración, definiendo el capital base invertido e imprimiendo las cantidades.

```python
class BootCampTask(QCAlgorithm):

    def Initialize(self):
        
        # seis meses
        self.SetStartDate(2020, 9, 1)
        self.SetEndDate(2021, 2, 1)
        
        #capital inicial invertido
        self.SetCash(1000000)

        # Select stocks and add in DataNormalizationMode = Raw
        self.nflx = self.AddEquity("ATVI", Resolution.Minute)
        self.nflx.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.dis = self.AddEquity("MSFT", Resolution.Minute)
        self.dis.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.fb = self.AddEquity("GOOG", Resolution.Minute)
        self.fb.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.tsla = self.AddEquity("IBM", Resolution.Minute)
        self.tsla.SetDataNormalizationMode(DataNormalizationMode.Raw)
        
        self.goog = self.AddEquity("ORCL", Resolution.Minute)
        self.goog.SetDataNormalizationMode(DataNormalizationMode.Raw)
        

    def OnData(self, data):
    # distribución regular de capital
        if not self.Portfolio.Invested:
            self.SetHoldings("ATVI", 0.2)
            self.SetHoldings("MSFT", 0.2)
            self.SetHoldings("GOOG", 0.2)
            self.SetHoldings("IBM", 0.2)
            self.SetHoldings("ORCL", 0.2)
            
            
            # impresión de datos
            self.Log("Quantity:" + str(self.Portfolio['ATVI'].Quantity));
            self.Log("Quantity:" + str(self.Portfolio['MSFT'].Quantity));
            self.Log("Quantity:" + str(self.Portfolio['GOOG'].Quantity));
            self.Log("Quantity:" + str(self.Portfolio['IBM'].Quantity));
            self.Log("Quantity:" + str(self.Portfolio['ORCL'].Quantity));
```
